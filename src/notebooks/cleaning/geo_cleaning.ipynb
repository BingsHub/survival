```python
import pandas as pd
from survival.utils import lower_case
```

```python
# column names for reference
columns = [
    'geonameid', 'name', 'asciiname', 'alternatenames', 'latitude', 'longitude',
    'feature_class', 'feature_code', 'country_code', 'cc2', 'admin1_code',
    'admin2_code', 'admin3_code', 'admin4_code', 'population', 'elevation',
    'dem', 'timezone', 'modification_date'
]
```

```python
# load data
cities = pd.read_csv('../../data/raw/cities500.txt', sep='\t', header=None, names=columns, usecols=['name', 'latitude', 'longitude', 'country_code'])
nl = pd.read_csv('../../data/raw/NL.txt', sep='\t', header=None, names=columns, usecols=['name', 'alternatenames', 'feature_code', 'latitude', 'longitude', 'country_code'])
countries = pd.read_csv('../../data/raw/allCountries.txt', sep='\t', header=None, names=columns, usecols=['name', 'feature_code', 'latitude', 'longitude', 'country_code'])
postalcode = pd.read_csv('../../data/raw/NLzipcode.txt', sep='\t', header=None, names=['country_code', 'pc_4', 'place_name', 'admin_name1', 'admin_code1', 'admin_name2', 'admin_code2', 'admin_name3', 'admin_code3', 'latitude', 'longitude', 'accuracy'], usecols=['country_code', 'pc_4', 'place_name', 'admin_name1', 'latitude', 'longitude'])
```

```python
lower_case(nl)
lower_case(cities)
lower_case(countries)
lower_case(postalcode)
```

```python
nl.to_csv('../../data/processed/nl.csv', index=False)
cities.to_csv('../../data/processed/cities.csv', index=False)
countries.to_csv('../../data/processed/countries.csv', index=False)
postalcode.to_csv('../../data/processed/postalcode.csv', index=False)
```
