```python
import pandas as pd

# import url
url = "https://nl.wikipedia.org/wiki/Lijst_van_voormalige_Nederlandse_gemeenten"


# read all tbles on the page
table = pd.read_html(url)

# select table (theres only one but still)
table = table[0]
```

```python
# get useful column names, rename, and change values to lower case
gemeenten = table[['Gemeente', 'Opgegaan in']]
gemeenten.columns = ['oude_naam', 'nieuwe_naam']
lower_case(gemeenten)
```

```python
# show all values of nieuwe_naam which contain a ';'
gemeenten[gemeenten['nieuwe_naam'].str.contains(';')]
```

```python
# export to csv
gemeenten.to_csv('../../data/processed/gemeenten.csv', index=False)
```
