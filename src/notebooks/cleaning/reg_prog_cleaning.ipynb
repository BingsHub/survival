```python
# import libraries
import pandas as pd
```

```python
# load data
reg = pd.read_csv('../../data/raw/operaballet_reg_prods.csv')
```

```python
# drop unnessecary columns
reg = reg.drop(columns=['season', 'artform'])
```

```python
reg['start_ticket_sales'] = pd.to_datetime(reg['start_ticket_sales'], format='mixed', dayfirst=True)
reg = reg.rename(columns={'start_ticket_sales': 'sale_start_date'})
```

```python
# save data
reg.to_csv('../../data/processed/operaballet_reg_prods_clean.csv', index=False)
```
