```python
# import libraries and custom modules
import pandas as pd
from survival.utils import lower_case
```

```python
# load data
income = pd.read_csv('../../data/raw/median-income-by-country-2024.csv')
```

```python
# select columns and rename for convenience
income = income[['country', 'MedianIncome_GDPPPP_IntDollars']]
income.columns = ['country_code', 'median_income']
```

```python
# change country_code names to country_code codes
from survival.constants import country_codes_income_dict

income['country_code'] = income['country_code'].map(country_codes_income_dict)
```

```python
# divide median income by 1000 and transform to lower case
income['median_income'] = income['median_income'] / 1000
lower_case(income)
```

```python
# export to csv
income.to_csv('../../data/processed/income.csv', index=False)
```
