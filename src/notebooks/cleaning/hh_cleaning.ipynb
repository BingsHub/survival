```python
# import libraries
import pandas as pd
```

```python
# load the data
hh = pd.read_csv('../../data/raw/huishoudinkomen2021.csv', sep=';')
```

```python
# retain only necessary colums
hh = hh[["Regio's" ,"Inkomen/Gemiddeld besteedbaar inkomen (1 000 euro)"]]

# rename columns
hh.columns = ['municipality', 'avg_income']
```

```python
# all string to lower case
def lower_case(data):
    for col in data.columns:
        if data[col].dtype == 'object':
            data[col] = data[col].str.lower()
    return data

hh = lower_case(hh)
```

```python
# change avg_income to a suitable format
hh['avg_income'] = hh['avg_income'].str.replace(',', '.').astype(float)
```

```python
replace_dict = {
    'nuenen, gerwen en nederwetten': 'nuenen',
    'beek (l.)': 'beek',
    'groningen (gemeente)': 'groningen',
    'hengelo (o.)': 'hengelo',
    'laren (nh.)': 'laren',
    "'s-gravenhage (gemeente)": 'den haag',
    'middelburg (z.)': 'middelburg',
    'rijswijk (zh.)': 'rijswijk',
    'stein (l.)': 'stein',
    'utrecht (gemeente)': 'utrecht'
}

for old_name, new_name in replace_dict.items():
    hh.loc[hh['municipality'] == old_name, 'municipality'] = new_name
```

```python
# to csv
hh.to_csv('../../data/processed/hh_clean.csv', index=False)
```
