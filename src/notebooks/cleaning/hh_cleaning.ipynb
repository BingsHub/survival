```python
# import libraries
import pandas as pd

# import customer modules
from survival.utils import lower_case
```

```python
# load the data
hh = pd.read_csv('../../data/raw/huishoudinkomen2021.csv', sep=';')
```

```python
# retain only necessary colums
hh = hh[["Regio's" ,"Inkomen/Mediaan besteedbaar inkomen (1 000 euro)"]]

# rename columns
hh.columns = ['municipality', 'median_income']

# change median_income to a suitable format
hh['median_income'] = hh['median_income'].str.replace(',', '.').astype(float)

# to lower case
hh = lower_case(hh)
```

```python
# fix the municipality names
from survival.constants import hh_replace_dict

for old_name, new_name in hh_replace_dict.items():
    hh.loc[hh['municipality'] == old_name, 'municipality'] = new_name
```
